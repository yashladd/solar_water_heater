```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd

# Assuming df is your dataframe
def plot_poa_global(df, month=None, day=None):
    plt.figure(figsize=(12, 6))

    # Plot for a specific day of a specific month
    if month is not None and day is not None:
        daily_df = df.loc[df.index.to_series().dt.month == month]
        daily_df = daily_df.loc[daily_df.index.to_series().dt.day == day]
        plt.plot(daily_df.index, daily_df['poa_global'])
        plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))
        plt.title(f'POA Global for {day}-{month}')

    # Plot for all days in a specific month
    elif month is not None:
        monthly_df = df.loc[df.index.to_series().dt.month == month]
        plt.plot(monthly_df.index, monthly_df['poa_global'])
        plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%d'))
        plt.title(f'POA Global for Month {month}')

    # Plot for the entire year with month on the x-axis
    else:
        plt.plot(df.index, df['poa_global'])
        plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b'))
        plt.title('POA Global for the Entire Year')

    # Common plot settings
    plt.xlabel('Time')
    plt.ylabel('POA Global')
    plt.grid(True)
    plt.legend(['POA Global'])
    plt.tight_layout()
    plt.show()

# Example usage:
# plot_poa_global(df, month=1, day=1) # For January 1st
# plot_poa_global(df, month=1) # For January
# plot_poa_global(df) # For the entire year
```

```python
plot_poa_global(simulation.environment_conditions)
```
